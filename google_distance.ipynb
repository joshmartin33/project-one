```python
import pandas as pd
import json
import requests

# Import API key
from api_keys import g_key
```

```python
# Store csv path and name
csv = ("Resources/Cleaned_VCE_results_by_school.csv")

# Read in csv file
school_df = pd.read_csv(csv) 
```

```python
# Set target city as Melbourne Australia
target_city = "Melbourne, Australia"

# params dictionary
params = {"address": target_city, "key": g_key}

# Build URL using the Google Maps API
base_url = "https://maps.googleapis.com/maps/api/geocode/json"

# Run request
mel_loca = requests.get(base_url, params=params).json()

# Adding collected data to lat and long variables
melbourne_lat = mel_loca["results"][0]["geometry"]["location"]["lat"]
melbourne_lon = mel_loca["results"][0]["geometry"]["location"]["lng"]

# Convert melbourne lat and long variables to usable string
origin = f"{melbourne_lat}+{melbourne_lon}"
```

```python
# Add column for Distance from city centre
school_df["Dist From Melbourne Centre (m)"] = ""

    # Use the school lat and long to identify distance from melbourne CBD
for index, row in school_df.iterrows():
        
        # Grab lat and long for each row and convert to usable string
        dest_lat = row["Latitude"]
        dest_lon = row["Longitude"]
        destination = f"{dest_lat}+{dest_lon}"
        

         # Use the units term to return distance metric data for each destination city
        target_url = f"https://maps.googleapis.com/maps/api/distancematrix/json?origins={origin}&destinations={destination}&key={g_key}"

         # make request and convert to json
        response = requests.get(target_url).json()

        # Adding collected data to school_df
        school_df.loc[index, "Dist From Melbourne Centre (m)"] = response["rows"][0]["elements"][0]["distance"]["value"]
```
